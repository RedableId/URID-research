# Idea

We use alphabet of size <= 40 and store each 3 symbols in 16 bits.
Alphabet:
- Not a symbol, zero means -- nothing to decode (1 used)
- 26 letters (27 used)
- 10 digits (37 used)
- Some other 2 symbols like `$`, `@`, `+`, `~`, `(`, `)`, etc (39 used)
- Separator (40 used)

So each 16-bit sub-vector $V_i$ encodes 3 symbols $s_i$ such way:

$V_i = s_0 + s_1 \cdot B + s_2 \cdot B^2$

# Density

```python
vector_lengths = [64, 128, 192, 256]

for vector_length in vector_lengths:
    triplets_count = vector_length // 16
    symbols_count = triplets_count * 3
    density = vector_length / symbols_count
    print(f"{vector_length}-bits vector contains {symbols_count} symbols, density {density:.3f}")
```
