# Idea

The minimal alphabet to distinguish tokens is 37 symbols:
- 26 letters
- 10 digits
- 1 separator

So each identifier is a 37-base radix number.

# Density

```python
import math

vector_lengths = [64, 128, 192, 256]

alphabet_size = 37
alphabet_size_bits = math.log2(alphabet_size)

for vector_length in vector_lengths:
    symbols = math.floor(vector_length / alphabet_size_bits)
    unused = vector_length - symbols * alphabet_size_bits
    density = vector_length / symbols
    print(f"{vector_length}-bits vector contains {symbols} symbols, {unused} bits are unused, density: {density:.2f}")
```
