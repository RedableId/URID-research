# Idea

Pack each symbol as 6-bit value.

Alphabet:
* `0-9` -- 10 symbols (10 used)
* `a-z` -- 26 symbols (36 used)
* `A-Z` -- 26 symbols (62 used)
* `Separator` -- 1 symbol (63 used)
* `End of string` -- 1 symbol (64 used)

# Density



```python
vector_lengths = [64, 128, 192, 256]

bits_per_symbol = 6

for vector_length in vector_lengths:
    symbols = vector_length // bits_per_symbol
    unused = vector_length - symbols * bits_per_symbol
    print(f"{vector_length}-bits vector contains {symbols} symbols, {unused} bits are unused")
```

# Operations

All operations are efficient due to bitwise operations:

- Encode ~ $O(L_{id})$ because we need to iterate through all symbols.
- Decode ~ $O(L_{id})$ because we need to iterate through all symbols.
- GetLength ~ $O(L_{id})$ because we need to iterate until we find the end of string symbol.
- GetSymbol ~ $O(1)$ because we can directly access the symbol by index.
- GetToken ~ $O(L_{id})$ because we need to iterate counting separators with index lower than requested and then iterate until next separator.
- AppendSymbol ~ $O(L_{id})$ because it requires GetLength.
- AppendToken ~ $O(L_{id})$ because it requires GetLength.
