# Idea

Pack each symbol as 5-bit value. Some letters and diggits are merged into one symbol.
Alphabet:
* `0-9` -- 10 symbols (10 used)
* `A-Z` except `I=1`, `O=0`, `S=5`, `Z=2` -- 22 symbols (32 used)
* To distinguish the end of word one more less obvious hack should be used:
    * `W=VV`, `T=7`, `A=4`, etc. -- now we can have one more symbol for separator

# Density

```python
vector_lengths = [64, 128, 192, 256]

bits_per_symbol = 5

for vector_length in vector_lengths:
    symbols = vector_length // bits_per_symbol
    unused = vector_length - symbols * bits_per_symbol
    print(f"{vector_length}-bits vector contains {symbols} symbols, {unused} bits are unused")
```

# Operations

All operations are efficient due to bitwise operations:

* `Encode` ~ $O(L_{id})$
* `Decode` ~ $O(L_{id})$
* `Get` ~ $O(1)$
* `Replace` ~ $O(1)$
* `Insert` ~ $O(L_{id})$
* `Remove` ~ $O(L_{id})$
